# On Determinants
__Working with square matricies.__

## Determinants I - Properties

### The Big Three

1. $\det I = 1$
2. Exchange rows $\to$ reverse sign of $\det$.

> Note that from $1$ and $2$ we know that  

$$
\det P = 
\begin{cases}
    1, &\text{if number of exchanges is even,} \\
    -1, &\text{if number of exchanges is odd.}
\end{cases}
$$
where $P$ is the permitation matrix.

3. The determinant is a linear function in one row, holding the other $n - 1$ rows constant.
    
    3.a
    $$
    \begin{vmatrix}
    ta & tb \\
    c & d
    \end{vmatrix} = t
        \begin{vmatrix}
        a & b \\
        c & d
        \end{vmatrix}
    $$
    3.b 
    $$
    \begin{vmatrix}
    a+a' & b+b' \\
    c & d
    \end{vmatrix} =
        \begin{vmatrix}
        a & b \\
        c & d
        \end{vmatrix} +
            \begin{vmatrix}
            a' & b' \\
            c & d
            \end{vmatrix}
    $$
    

### The remaining properties are all derived form _The Big Three_.

4. If two rows are equal, then the $\det$ is zero.

> Using (2) exchange those rows, which gives the same matrix. Now since the matrix is the same, the determinant must also be the same; but buy (2) the determinant should be different (k if it was -k, -k if it was k). This can only hold if $\det = 0$.

5. Elimination property; subtracting $l \times R_i$ from $R_j$ leaves the determinant unchanged.

>$$
\begin{align}
    \begin{vmatrix}
    a & b \\
    c-la & d-lb
    \end{vmatrix} &=
        \begin{vmatrix}
        a & b \\
        c & d
        \end{vmatrix} +
        \begin{vmatrix}
        a & b \\
        -la & -lb
        \end{vmatrix} \quad \text{by (3b)}\\
        &=
        \begin{vmatrix}
        a & b \\
        c & d
        \end{vmatrix} + (-l)
        \begin{vmatrix}
        a & b \\
        a & b
        \end{vmatrix} \quad \text{by (3a)}\\
        &=
        \begin{vmatrix}
        a & b \\
        c & d
        \end{vmatrix} \quad \text{by (4)}
\end{align}
$$

6. A row of zero's implies that the determinate is zero.

> Use $(3a)$ with (t = 0).

7. The determinant of an upper triangular matrix, U, is the product of its pivots.

> Using the elimination property $(5)$, reduce $U$ to $D$, a diagonal matrix. Then using $(3a)$ we can factor out the pivots $(d_1, d_2, \cdots, d_n)$ leaving the identity multiplied by their products. By $(1)$ we know that $\det I = 1$, thus $\det D = (d_1)(d_2) \cdots (d_n)$.

_Note:_ If some $d_i = 0$, then using $(5)$ we can zero out a row, then by $(6)$ the determinant is zero so the property still holds.

8. $\det A = 0$ precisely when $A$ is singular (so the determinant can function as a test for invertibility).

> Suppose $A$ is singular, then by elimination we get a row of zero which means the determinant is zero.

And equivalently $\det A \neq 0$ when $A$ is invertible.

> If $A$ is invertible, then by $(7)$ $A \to U \to D$ with determinant $(d_1)(d_2) \cdots (d_n)$ where the $d's$ are all non-zero, and so the determinant is non-zero.

9. $\det AB = (\det A)(\det B)$.

> What is $\det A^{-1}$? We know $A^{-1}A = 1$, so $\det A^{-1}A = (\det A^{-1})(\det A) = \det I = 1$. Now if $(\det A^{-1})(\det A) = 1$, then
$$
\det A^{-1} = \frac{1}{\det A}.
$$

10. $\det A^{T} = \det A$.

> Extends all previous properties to columns. Suppose we have $A \to LU$, so

$$
\begin{align}
|A^T| &= |A| \\
|(LU)^T| &= |LU| \\
|U^T L^T| &= |LU| \\
|U^T||L^T| &= |L||U|.
\end{align}
$$

These can be reduced to diagonal matricies whose determinant is the product of the pivots, and transposing does not affect them. Thus $\det A^{T} = \det A$ must hold.

## Determinants II - Formula and Computation

First consider the $2 \times 2$ case

$$
\begin{align}
\begin{vmatrix}
a & b \\
c & d
\end{vmatrix} 
    &= \begin{vmatrix}
    a & 0 \\
    c & d
    \end{vmatrix} + 
        \begin{vmatrix}
        0 & b \\
        c & d
        \end{vmatrix} \\
    &= \begin{vmatrix}
    a & 0 \\
    c & 0
    \end{vmatrix} + 
        \begin{vmatrix}
        a & 0 \\
        0 & d
        \end{vmatrix} +
            \begin{vmatrix}
            0 & b \\
            c & 0
            \end{vmatrix} + 
                \begin{vmatrix}
                0 & b \\
                0 & d
                \end{vmatrix} \\
    &= \begin{vmatrix}
    a & 0 \\
    0 & d
    \end{vmatrix} +
        \begin{vmatrix}
        0 & b \\
        c & 0
        \end{vmatrix} \\
    &= ad - bc
\end{align}
$$

The $3 \times 3$ case is approched in a similar way were the determinant is expanded and all the determinats with zero rows or coluomns are zero. Only the determinants with one zero in each row and coulmn will remain.
