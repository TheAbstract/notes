The central limit theorem (CLT) states that as the number of samples you take from a population increase, the means of those samples will approximate a normal distribution.

```python
from random import choice, randint
from statistics import mean
import matplotlib.pyplot as plt
```

```python
n = 1000
data = [randint(0, 100) for _ in range(n)]
```

```python
plt.hist(data, bins='auto')
plt.title("Distribution not normal")
plt.show()
```

Take samples with a sample size of 5, plot the frequency of the mean of each sample.

```python
n = 10_000
means = []

for _ in range(n):
    sample = [choice(data) for _ in range(5)]
    means.append(mean(sample))
```

```python
plt.hist(means, bins='auto')
plt.title("See, it's normal!")
plt.show()
```

This holds for almost any distribution, continuous or discrete, beta, gamma or uniform.
