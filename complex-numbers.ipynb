# On Complex Numbers

## 1. The imaginary unit $i$

The imaginary unit $i$ is by definition a number such that $i^{2} = -1$.

### Powers of $i$

$$
\begin{align}
i^{0} &= 1 \quad \text{by exponent properties} \\
i^{1} &= i \quad \text{by exponent properties} \\
i^{2} &= -1 \quad \text{by definition} \\
&\text{(Note that all remaing powers of $i$ can be deduced from the first three.)} \\
i^{3} &= i^{2} \cdot i = -1 \cdot i = -i \\
i^{4} &= i^{3} \cdot i = -i \cdot i = 1 \\
i^{5} &= i^{4} \cdot i = 1 \cdot i = i \\
i^{6} &= i^{5} \cdot i = i \cdot i = -1 \\
i^{7} &= i^{6} \cdot i = -1 \cdot i = -i
\end{align}
$$

### $i$ to any exponent

$$
\begin{align}
i^{100} &= i^{4 \cdot 24} = (i^{4})^{25} = 1^{25} = 1. \\
i^{501} &= i^{500} \cdot i^{1} = (i^{4})^{125} \cdot i = 1^{125} \cdot i = i.
\end{align}
$$

This is possible because $i^{4k} = 1$, since $i^{4k} = (i^{4})^{k} = 1^{k} = 1, k ≥ 0$.

$$
\begin{align}
i^{99} &= i^{96} \cdot i^{3} = (i^{4})^{16} \cdot -i = 1^{16} \cdot -i = -i, \\
i^{38} &= i^{36} \cdot i^{2} = 1 \cdot i^{2} = -1.
\end{align}
$$

> Idea is to express the exponent as some multiple of four and some other stuff. Then the multiple of four will become $1$, and the rest should be easy to compute.

## 2. imaginary roots of negative numbers

Simplify: $\sqrt{-52}$.

First, $\sqrt{(-1)(52)} = \sqrt{-1}\cdot \sqrt{52}$.

> Note that $\sqrt{ab} = \sqrt{a}\cdot \sqrt{b}$ only if $a$ and $b$ are not both negative. For example $\sqrt{52} = \sqrt{(-1)(-52)} \neq \sqrt{-1}\cdot \sqrt{-52}$.

Continuing from above

$$
\begin{align}
\sqrt{-52} &= \sqrt{(-1)(52)} = \sqrt{-1}\cdot \sqrt{52} \\
    &= i \cdot \sqrt{52} \\
    &= i \cdot \sqrt{4 \cdot 13} \\
    &= i \cdot \sqrt{4} \cdot \sqrt{13} \\
    &= i \cdot 2 \sqrt{13} \\
    &= 2 \sqrt{13} i
\end{align}
$$

## 3. Complex numbers

Consider $z = 5 + 3i$, this whole thing is a complex number, with $Re(z) = 5$ and $Im(z) = 3$. So we see that comlex numbers are combinations of real numbers and multiples of the imaginary unit.

### Operations

#### Addition
Let $a = 3 + 2i$ and $b = -1 -3i$, then $a + b = 2 - i$.

#### Multiplication
$$
\begin{align}
(1 - 3i)(2 + 5i) &= 2(1 - 3i) + 5i(1 - 3i) \\
    &= 2 - 6i + 5i - 15i^{2} \\
    &= 2 - 6i + 5i + 15 \\
    &= 17 - i.
\end{align}
$$

#### Conjugates
If $z = a + bi$, then $\bar{z} = a - bi$.

> Note: $z + \bar{z} = 2a$ and $z \cdot \bar{z} = a^2 + b^2$.

__Dividing complex numbers using conjugates__

Consider,
$$
\frac{1 + 2i}{4 - 5i}
$$
using the conjugate of the denominator we get

$$
\begin{align}
\frac{1 + 2i}{4 - 5i} \cdot \frac{4 + 5i}{4 + 5i} &= \frac{4 + 5i + 8i + 10i^{2}}{4^{2} - (5i)^{2}} \\
    &= \frac{4 - 10 + 13i}{16 + 25} \\
    &= \frac{-6 + 13i}{41} \\
    &= \frac{-6}{41} + \frac{13i}{41}.
\end{align}
$$


## $\times$ and $÷$ in polar form



For complex numbers $z_1$ and $z_2$ (whith radii and angles $r_1, r_2$ and $\theta_1, \theta_2$)
$$
\begin{align}
z_1 &= r_1\cos\theta_1 + i\sin\theta_1 \\
z_2 &= r_2\cos\theta_2 + i\sin\theta_2
\end{align}
$$

Multiplication:
- The radius of $z_1 \cdot z_2$ is $r_1 \cdot r_2$
- The angle of $z_1 \cdot z_2$ is $\theta_1 + \theta_2$

So
$$
z_1z_2 = r_1r_2[\cos(\theta_1 + \theta_2) + i\sin(\theta_1 + \theta_2)].
$$

Division:
- The radius of $\frac{z_1}{z_2}$ is $\frac{r_1}{r_2}$
- The angle of $\frac{z_1}{z_2}$ is $\theta_1 - \theta_2$

So
$$
\frac{z_1}{z_2} = \frac{r_1}{r_2} [\cos(\theta_1 - \theta_2) + i\sin(\theta_1 - \theta_2)].
$$

## 4. Basic complex analysis

Let $z = a + bi$, the vector $z$ may be represented in an argand diagram using coordinates $(a, b)$. The point $(a, b)$ may also be represented by specifing an angle $\phi$ and the magnitude of the vector $ r = |z|$;
$$
r = |z| = \sqrt{a^{2} + b^{2}}, \quad \phi = \arctan \frac{a}{b}.
$$

Given $\phi, r$ we need to find $a, b$;
we know that $\cos \phi = \frac{a}{r}$, thus $a = r \cos \phi$, and similarly $\sin \phi = \frac{b}{r}$, thus $b = r \sin \phi$. From this we get 
$$
\begin{align}
z &= r \cos \phi + r \sin \phi i \\
  &= r (\cos \phi + i\sin \phi) \quad \text{(polar form)} \\
  &= r e^{i \phi} \quad \text{(exponential form)}.
\end{align}
$$

Given $z = \frac{\sqrt{3}}{2} + \frac{1}{2} i$, find $r, \phi$;
now $r = \sqrt{\frac{3}{4} + \frac{1}{4}} = 1$, and
$$
\begin{align}
\phi &= \arctan \frac{b}{a} \\
    &= \arctan \frac{1}{2} \cdot \frac{2}{\sqrt{3}} \\
    &= \arctan \frac{1}{\sqrt{3}} \\
\end{align}
$$

## 5. Using exponential form of complex numbers

__I__
$$
\begin{align}
\frac{7 (\cos \frac{7\pi}{6} + i \sin \frac{7\pi}{6})}{\cos \frac{7\pi}{4} + i \sin \frac{7\pi}{4}}
    &= \frac{7e^{\frac{7\pi}{6}i}}{e^{\frac{7\pi}{4}i}} \\
    &= 7e^{\frac{7\pi}{6}i - \frac{7\pi}{4}i} \\
    &= 7e^{\frac{14\pi i - 21 \pi i}{12}} \\
    &= 7e^{\frac{-7 \pi i}{12}} \\
    &= 7(\cos \frac{-7 \pi}{12} + i \sin \frac{-7 \pi}{12}).
\end{align}
$$

__II__
$$
\begin{align}
(\cos (\frac{2}{3} \pi) + i \sin (\frac{2}{3} \pi))^{20} &= (e^{\frac{2}{3} \pi i})^{20} \\
    &= e^{\frac{2}{3} \pi i \cdot 20} \\
    &= e^{\frac{40}{3} \pi i}
\end{align}
$$

__Find the roots of $x^3 = 1$.__

Let $z = 1$, then $|z| = 1$, with Arg$(z) = 0, 2\pi, 4\pi, 6\pi, \dots$. 

Thus
$$
1 = e^{0i} = e^{2\pi i} = e^{4\pi i} = \dots
$$

$$
\begin{align}
x^{3} &= 1, \quad e^{2\pi i}, \quad e^{4\pi i} \\
x &= (1)^{-3}, \quad (e^{2\pi i})^{-3}, \quad (e^{4\pi i})^{-3} \\
  &= 1, \quad e^{\frac{2\pi i}{3}}, \quad e^{\frac{4\pi i}{3}}
\end{align}
$$

Note that $r = 1, 1, 1$ and $\phi = 1, \frac{2\pi}{3}, \frac{4\pi}{3}$.
